```python
path='UrbanSound8K/UrbanSound8K/audio/fold1\\101415-3-0-2.wav'
```

```python
import librosa
def create_spec(y):
    spec=librosa.feature.melspectrogram(y=y)
    spec_conv=librosa.amplitude_to_db(spec,ref=np.max)
    return spec_conv

y, sr = librosa.load(path)

create_spec(y)
```

```python
import uuid
import cv2
def saver():
    for key in getList(classes.keys()):
        for i in range(len(classes[key])):
            sample=classes[key][i]
            y, sr = librosa.load(sample)
            spec=create_spec(y)
            
            plt.imsave(f'C:/Users/mehmu/Desktop/GlobalAI/preprocessedData/{key}/{str(uuid.uuid4())}.jpg',spec)
saver()
```
